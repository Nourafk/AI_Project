```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import warnings 
warnings.filterwarnings('ignore')
```

```python
#Importing the dataset with pandas
dataset = pd.read_csv('pre_advertisment(after2).csv') 
```

```python
#Declare feature vector and target variable
feature_cols =['Daily Time Spent on Site','Age','Area Income','Internet usage bucket','Male','Daily Internet Usage'] 
x=dataset[feature_cols] #feature columns
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss =[]
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter =300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the graph to visualize the Elbow Method to find the optimal number of cluster
plt.plot(range(1,11),wcss) 
plt.title('The Elbow Method') 
plt.xlabel('Number of clusters') 
plt.ylabel('WCSS')
plt.show()
```

```python
#We utilized the elbow approach to get the ideal number of K for our dataset and to determine the range of the optimal Ks that should be picked. 
#Therefore, the three alternatives we will employ for the k-mean clustering method are K=2, K=3, and K=4.
```

```python
#Applying KMeans to the dataset with 2 as the number of cluster
clustering_kmeans = KMeans(n_clusters=2)
x['clusters'] = clustering_kmeans.fit_predict(x)
```

```python
# Here we have applied a principal component analysis because we have more then 2 variables and it helps
#us with visulizing the cluster
pca_num_components = 2 
reduced_data = PCA(n_components=pca_num_components).fit_transform(x)
results = pd.DataFrame(reduced_data, columns = ['pca1', 'pca2'])
sns.scatterplot(x="pca1",y="pca2", hue= x['clusters'], data=results )
plt.title('K-means Clustering')
plt.show()
```

```python
# here we will calculate the accuracy of K=2 
labels = clustering_kmeans.labels_
y= dataset['Clicked on Ad'].astype(object)
# check how many of the samples were correctly labeled
correct_labels = sum(y == labels)

print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
```

```python
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python
#evaluate the result of the chosen K by using the silhouette coefficient.
#Import
from sklearn.metrics import silhouette_score
#Calculate the score
y_kmeans = kmeans.fit_predict(x)
score = silhouette_score(x,y_kmeans)
#Print it
print(score)
```

```python
#A low SSE value means that the centroids accurately represent the data points inside each cluster and that the data points are well-clustered.
#calculate sse
clustering_kmeans.inertia_
```

```python
#checking the Coordinates of cluster centers.
clustering_kmeans.cluster_centers_
```
